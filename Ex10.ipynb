```python
from transformers import pipeline
import gradio as gr
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
def summarize_text(text):
    if not text.strip():
        return "Please enter some text to summarize."
    result = summarizer(text, max_length=120, min_length=30, do_sample=False)
    return result[0]['summary_text']
```

```python
with gr.Blocks() as demo:
    gr.Markdown("## üß† Text Summarizer using Pre-trained LLM")
    txt = gr.Textbox(lines=8, label="Enter text")
    out = gr.Textbox(label="Summary")
    btn = gr.Button("Summarize")
    btn.click(summarize_text, txt, out)
```

```python
demo.launch(share=True, debug=True)
```
